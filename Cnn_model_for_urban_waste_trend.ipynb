```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Read the CSV file into a pandas DataFrame
df = pd.read_csv("/content/drive/MyDrive/singapore waste managenet/waste_management_Singapore_2003_to_2020.csv")

# Step 2: Clean the dataset
# Remove commas from numeric columns and convert them to numeric data types
numeric_columns = ['Disposed_(tonne)', 'Recycled_tonne', 'Generated_tonne']
df[numeric_columns] = df[numeric_columns].replace({',': ''}, regex=True).astype(float)

# Step 3: Plot graphs for data visualization
# Let's plot a bar chart for each year showing the amount of waste disposed and recycled
years = df['Year'].unique()
for year in years:
    data_year = df[df['Year'] == year]
    plt.figure(figsize=(10, 6))
    plt.bar(data_year['Waste Type'], data_year['Disposed_(tonne)'], label='Disposed')
    plt.bar(data_year['Waste Type'], data_year['Recycled_tonne'], label='Recycled')
    plt.xlabel('Waste Type')
    plt.ylabel('Amount (tonnes)')
    plt.title(f'Waste Disposed vs Recycled in {year}')
    plt.xticks(rotation=90)
    plt.legend()
    plt.tight_layout()
    plt.show()


```

```python
import pandas as pd

# Load the dataset
file_path = '/content/drive/MyDrive/singapore waste managenet/waste_management_Singapore_2003_to_2020.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the dataset
data.head()
```

```python
import numpy as np

# Cleaning data
data['Disposed_(tonne)'] = data['Disposed_(tonne)'].str.replace(',', '').astype(float)
data['Recycled_tonne'] = data['Recycled_tonne'].str.replace(',', '').astype(float)
data['Generated_tonne'] = data['Generated_tonne'].str.replace(',', '').astype(float)
data['Recycling Rate'] = data['Recycling Rate'].str.rstrip('%').astype(float) / 100

# Summarizing data yearly
yearly_summary = data.groupby('Year').agg({
    'Disposed_(tonne)': 'sum',
    'Recycled_tonne': 'sum',
    'Generated_tonne': 'sum',
    'Recycling Rate': 'mean'
}).reset_index()

yearly_summary.head()
```

```python
# @title Recycled_tonne vs Generated_tonne

from matplotlib import pyplot as plt
yearly_summary.plot(kind='scatter', x='Recycled_tonne', y='Generated_tonne', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import Conv1D, MaxPooling1D, Flatten, Dense, Dropout
from keras.optimizers import Adam
from keras import regularizers

# Preparing data for the model
data_input = yearly_summary[['Year', 'Generated_tonne']].set_index('Year')
scaler = MinMaxScaler()
data_scaled = scaler.fit_transform(data_input)

# Generating sequences for training
def create_sequences(data, n_steps):
    X, y = [], []
    for i in range(n_steps, len(data)):
        X.append(data[i-n_steps:i])
        y.append(data[i])
    return np.array(X), np.array(y)

n_steps = 3
X, y = create_sequences(data_scaled, n_steps)

# Split data into train and test
split_idx = int(0.8 * len(X))
X_train, X_test = X[:split_idx], X[split_idx:]
y_train, y_test = y[:split_idx], y[split_idx:]

# Reshape from [samples, timesteps] to [samples, timesteps, features]
n_features = 1
X_train = X_train.reshape((X_train.shape[0], X_train.shape[1], n_features))
X_test = X_test.reshape((X_test.shape[0], X_test.shape[1], n_features))

# Define the model with regularization and dropout
model = Sequential([
    Conv1D(filters=64, kernel_size=2, activation='relu', input_shape=(n_steps, n_features)),
    MaxPooling1D(pool_size=2),
    Flatten(),
    Dense(50, activation='relu', kernel_regularizer=regularizers.l2(0.01)),
    Dropout(0.2),
    Dense(1)
])
model.compile(optimizer=Adam(learning_rate=0.01), loss='mse', metrics=['accuracy'])

#model.compile(optimizer=Adam(learning_rate=0.01), loss='mse')

# Train the model
history = model.fit(X_train, y_train, epochs=200, verbose=0, validation_data=(X_test, y_test))

# Training history plot
plt.figure(figsize=(10, 5))
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss Over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()

```

```python
# Predict future years
future_steps = 10
last_input = data_scaled[-n_steps:]
future_predictions = []

for i in range(future_steps):
    # Reshape last input to fit model input shape
    last_input_reshaped = last_input.reshape((1, n_steps, n_features))
    # Predict the next step
    next_step = model.predict(last_input_reshaped)
    # Append the prediction
    future_predictions.append(next_step)
    # Update last input using the prediction
    last_input = np.append(last_input[1:], next_step, axis=0)

# Inverse transform the predictions
future_predictions = np.array(future_predictions).reshape(-1, 1)
future_predictions = scaler.inverse_transform(future_predictions)
future_years = np.arange(2021, 2031)

# Plotting future predictions
plt.figure(figsize=(10, 5))
plt.plot(yearly_summary['Year'], yearly_summary['Generated_tonne'], label='Historical Data')
plt.plot(future_years, future_predictions, label='Predicted Data', linestyle='--')
plt.title('Future Waste Generation Predictions (2021-2030)')
plt.xlabel('Year')
plt.ylabel('Total Waste Generated (tonne)')
plt.legend()
plt.show()
```
